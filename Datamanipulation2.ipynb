###GROUPING AND AGGREGATION
We group our data by aggregation

```python
grouped=df.groupby("department")
grouped
```

```python
df.groupby("department")["salary"].sum()
```

```python
df.groupby("department")["salary"].mean()
```

```python
df.groupby('department').agg({
    'salary':['sum','mean','count','min','std','var','median'],
    })
```

```python
df.groupby('department').agg({'age':['sum','max','mean','var','count','median','min'],
})
```
